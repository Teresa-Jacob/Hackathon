```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import sklearn
```

```python
train_data = pd.read_csv('Train_data.csv', parse_dates=[0], dayfirst=True)
test_data = pd.read_csv('Test_data.csv', parse_dates=[0], dayfirst=True)
```

```python
train_data.head(1000)
```

```python
dateSelect = '28-02-2021'  
train_data['dates'] = dateSelect 
```

```python
dateSelect1 = '15-03-2021'  
test_data['dates'] = dateSelect1
```

```python
train_data['dates'] =  pd.to_datetime(train_data['dates'])
test_data['dates'] =  pd.to_datetime(test_data['dates'])
```

```python
train_data['days']= train_data['dates']-train_data['date']
train_data['day']=train_data['days'].dt.days
test_data['days']= test_data['dates']-test_data['date']
test_data['day']=test_data['days'].dt.days
```

```python
train_data.head(1000)
```

```python
test_data.head(3000)
```

```python
train_data.drop(['date', 'dates', 'days', 'campaign'], axis = 1, inplace=True)
test_data.drop(['date', 'dates', 'days', 'campaign'], axis = 1, inplace=True)
```

```python
train_data['adgroup'] = train_data['adgroup'].str.replace('adgroup ','')
test_data['adgroup'] = test_data['adgroup'].str.replace('adgroup ','')
```

```python
a= pd.get_dummies(train_data.adgroup, prefix='adgroup')
b= pd.get_dummies(test_data.adgroup, prefix='adgroup')
```

```python
train_data = pd.concat([a, train_data], axis=1)
del train_data['adgroup']
test_data = pd.concat([b, test_data], axis=1)
del test_data['adgroup']
```

```python
train_data['ad'] = train_data['ad'].str.replace('ad ','')
test_data['ad'] = test_data['ad'].str.replace('ad ','')
```

```python
train_data['ad'] =train_data['ad'].astype(int)
test_data['ad'] =test_data['ad'].astype(int)
```

```python
for i, row in train_data.iterrows():
    if (row['adgroup_1']==1):
        x=row['ad']
        train_data.at[i,'adgroup_1']=x
    if (row['adgroup_2']==1):
        x1=row['ad']
        train_data.at[i,'adgroup_2']=x1
    if (row['adgroup_3']==1):
        x2=row['ad']
        train_data.at[i,'adgroup_3']=x2
    if (row['adgroup_4']==1):
        x3=row['ad']
        train_data.at[i,'adgroup_4']=x3 
```

```python
for i, row in test_data.iterrows():
    if (row['adgroup_1']==1):
        x4=row['ad']
        test_data.at[i,'adgroup_1']=x4
    if (row['adgroup_2']==1):
        x5=row['ad']
        test_data.at[i,'adgroup_2']=x5
    if (row['adgroup_3']==1):
        x6=row['ad']
        test_data.at[i,'adgroup_3']=x6
    if (row['adgroup_4']==1):
        x7=row['ad']
        test_data.at[i,'adgroup_4']=x7 
```

```python
del train_data['ad']
del test_data['ad']
```
